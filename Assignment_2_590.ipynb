```python
x=2;
eta=0.001
eps=0.000001
delx=1
max_iter=10000
iter=0
def deriv(x):
  xderive=4*x*x*x+6*x
  return xderive
```

```python
while abs(delx)>eps and iter<max_iter:
    prevx=x;
    delx=-eta*deriv(prevx)
    x=x+delx
    iter=iter+1
print("Value of loacl minimum occurs",round(x))
obj=x*x*x*x+3*x*x+10
print("Value of objective function = ",round(obj))
```
